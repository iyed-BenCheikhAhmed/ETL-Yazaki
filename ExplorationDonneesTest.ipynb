# Importation des données après nettoyage

```python
# Importer les fonctions ETL (extraction + transformation)
from etl.extract import extract_charges_telephoniques, extract_charges_impression
from etl.transform import transform_charges_telephoniques, transform_charges_impression
```

```python
import pandas as pd

# Extraire et transformer les ChargesTelephoniques
ChargesTelephoniques = extract_charges_telephoniques()
ChargesTelephoniques = transform_charges_telephoniques(ChargesTelephoniques)

print(f"ChargesTelephoniques après nettoyage : {ChargesTelephoniques.shape}")
ChargesTelephoniques.head(10)
```

```python
# Extraire et transformer les ChargesImpression
ChargesImpression = extract_charges_impression()
ChargesImpression = transform_charges_impression(ChargesImpression)

print(f"ChargesImpression après nettoyage : {ChargesImpression.shape}")
ChargesImpression.head()
```

```python
# Détecter les employés ayant plus d'un rôle distinct (erreur de saisie probable)
role_counts = ChargesTelephoniques.groupby(['CodeEmployee', 'NomRole']).size().reset_index(name='NbOccurrences')
multi_role = role_counts.groupby('CodeEmployee').filter(lambda x: len(x) > 1)

print(f"Employés avec plusieurs rôles distincts : {multi_role['CodeEmployee'].nunique()}\n")
multi_role.sort_values(['CodeEmployee', 'NbOccurrences'], ascending=[True, False])
```

```python
# Détecter les employés ayant plus d'un numéro de téléphone distinct
phone_counts = ChargesTelephoniques.groupby(['CodeEmployee', 'NumeroTel']).size().reset_index(name='NbOccurrences')
multi_phone = phone_counts.groupby('CodeEmployee').filter(lambda x: len(x) > 1)

print(f"Employés avec plusieurs numéros distincts : {multi_phone['CodeEmployee'].nunique()}\n")
multi_phone.sort_values(['CodeEmployee', 'NbOccurrences'], ascending=[True, False])
```

```python
# Cohérence CodeDepartement / NomDepartement / NomResponsable
dept_info = (
    ChargesTelephoniques
    .groupby('CodeDepartement')
    .agg(
        NomDepartement=('NomDepartement', 'unique'),
        NbNoms=('NomDepartement', 'nunique'),
        NomResponsable=('NomResponsable', 'unique'),
        NbResponsables=('NomResponsable', 'nunique'),
    )
)

# Signaler les incohérences
incoherents = dept_info[(dept_info['NbNoms'] > 1) | (dept_info['NbResponsables'] > 1)]
print(f"Départements incohérents : {len(incoherents)}\n")
dept_info
```

```python
# Détecter les employés affectés à plus d'un département
dept_counts = ChargesTelephoniques.groupby(['CodeEmployee', 'NomDepartement']).size().reset_index(name='NbOccurrences')
multi_dept = dept_counts.groupby('CodeEmployee').filter(lambda x: len(x) > 1)

print(f"Employés avec plusieurs départements distincts : {multi_dept['CodeEmployee'].nunique()}\n")
multi_dept.sort_values(['CodeEmployee', 'NbOccurrences'], ascending=[True, False])
```

```python
# coherence CodeDepartement / NomDepartement / NomResponsable
dept_info_impression = (
    ChargesImpression
    .groupby('CodeDepartement')
    .agg(
        NomDepartement=('NomDepartement', 'unique'),
        NbNoms=('NomDepartement', 'nunique'),
        NomResponsable=('NomResponsable', 'unique'),
        NbResponsables=('NomResponsable', 'nunique'),
    )
)
# Signaler les incohérences
incoherents_impression = dept_info_impression[(dept_info_impression['NbNoms'] > 1) | (dept_info_impression['NbResponsables'] > 1)]
print(f"Départements incohérents : {len(incoherents_impression)}\n")
dept_info_impression
```

```python
# coherence CouleurImpression / CodeDetailImpression / Format
impression_info = (
    ChargesImpression
    .groupby('CouleurImpression')
    .agg(
        CodeDetailImpression=('CodeDetailImpression', 'unique'),
        NbCodes=('CodeDetailImpression', 'nunique'),
        Format=('Format', 'unique'),
        NbFormats=('Format', 'nunique'),
    )
)
# Signaler les incohérences
incoherents_impression_detail = impression_info[(impression_info['NbCodes'] > 1) | (impression_info['NbFormats'] > 1)]
print(f"Couleurs d'impression incohérentes : {len(incoherents_impression_detail)}\n")
impression_info
```

```python
# coherence CodeDetailImpression / CoutUnitaire
detail_info = (
    ChargesImpression
    .groupby('CodeDetailImpression')
    .agg(
        CoutUnitaire=('CoutUnitaire', 'unique'),
        NbCouts=('CoutUnitaire', 'nunique'),
    )
)
# Signaler les incohérences
incoherents_detail_cout = detail_info[detail_info['NbCouts'] > 1]
print(f"Codes de détail d'impression incohérents : {len(incoherents_detail_cout)}\n")
detail_info
```
